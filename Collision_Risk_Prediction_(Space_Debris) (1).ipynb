```python
from skyfield.api import load, EarthSatellite
from sklearn.ensemble import RandomForestClassifier
```

```python
data = {
    "distance_km":[5,10,15, 20,3],
    "rel_vel_km_s":[10,5, 6,3,15],
    "inclination_diff":[0.5,1.0,0.3,2.0,0.1],
    "risk": [1,0,1,0,1]
}
```

```python
df_2 = pd.DataFrame(data)
```

```python
df_2
```

```python
def train_risk_model():
  x = df_2[["distance_km","rel_vel_km_s","inclination_diff"]]
  y =df_2["risk"]
  model = RandomForestClassifier(n_estimators = 100, random_state=42)
  model.fit(x,y)
  return model
```

```python
ts = load.timescale()
t = ts.utc(2025, 3, 10,12)
sat1 = EarthSatellite("1 58990U 24033A   25069.85878478 -.00001735  00000-0 -17300-4 0 4242",
                      "2 58990 000.0390 256.4286 0002061 091.7602 211.7872 01.00271532  1555", "INSAT-3DS", ts)

debris = EarthSatellite("1 61994U 16053L   25069.03483903  .00000610  00000-0  61000-5 0  143",
                        "2 61994 000.3881 098.6600 0024969 193.0199 068.0215 01.00490959   143", "INTELSAT 33E DEB", ts)
```

```python
# Calculate features
pos1, pos2 = sat1.at(t).position.km, debris.at(t).position.km   # position vectors (x, y, z in km) for both objects at time t

distance_km = ((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2 + (pos1[2] - pos2[2])**2)**0.5   # Euclidean dist formula to calculate separation between  satellite and debris

vel1, vel2 = sat1.at(t).velocity.km_per_s, debris.at(t).velocity.km_per_s                      # diff in vel vectors
rel_vel_km_s = ((vel1[0] - vel2[0])**2 + (vel1[1] - vel2[1])**2 + (vel1[2] - vel2[2])**2)**0.5 # rel vel diff
```

```python
incl_1 = sat1.model.inclo
incl_2 = debris.model.inclo

inclination_diff = abs(incl_1 - incl_2)
```

```python
# Predict risk

model = train_risk_model()

features = [[distance_km, rel_vel_km_s, inclination_diff]]
risk = model.predict(features)[0]

print("Predicted Risk Level:", "High" if risk == 1 else "Low")
```

##**fetching the raw TLE data for the objects in dataset from an external source (like Space-Track or Celestrak) and then adjust or map those TLEs to match the objects in dataset**

```python
# map specific columns from your dataset to the fields in TLE Line 1 and Line 2
# dataset contains derived orbital elements (e.g., MEAN_MOTION, ECCENTRICITY, etc.)
# it does not include the raw TLE lines (line 1 and line 2) needed to create EarthSatellite objects in skyfield
```

```python
# hese raw TLE lines are essential for calculating the 3D positions of satellites and determining their distances
```

```python
# we reconstruct the tle lines for each object using the orbital elements provided in the dataset
# The TLE format is standardized, and we can map the columns in your dataset to the fields in a TLE
```

```python
# Line 1: Contains metadata like the satellite number, epoch, drag terms (BSTAR, MEAN_MOTION_DOT, MEAN_MOTION_DDOT), etc.
```

```python
# Line 2: Contains the orbital elements (INCLINATION, RA_OF_ASC_NODE, ECCENTRICITY, ARG_OF_PERICENTER, MEAN_ANOMALY, MEAN_MOTION, REV_AT_EPOCH).
```

```python
from datetime import datetime

def compute_checksum(line):
    """Compute the TLE checksum for a line."""
    checksum = 0
    for char in line[:-1]:
        if char.isdigit():
            checksum += int(char)
        elif char == '-':
            checksum += 1
    return checksum % 10

def format_epoch_to_tle(epoch_str):
    """Convert EPOCH (e.g., 2025-03-10T20:36:38.655936) to TLE format (YYDDD.DDDDDDDD)."""
    dt = datetime.strptime(epoch_str, "%Y-%m-%dT%H:%M:%S.%f")
    year = dt.year % 100  # Last two digits of year (e.g., 25 for 2025)
    day_of_year = dt.timetuple().tm_yday  # Day of year (1-366)
    fractional_day = (dt.hour * 3600 + dt.minute * 60 + dt.second + dt.microsecond / 1e6) / 86400
    day_str = f"{day_of_year:03d}.{fractional_day:08f}"[:12]  # YYDDD.DDDDDDDD
    return f"{year:02d}{day_str}"

```

```python
def reconstruct_tle(row):
    """Reconstruct TLE lines from a row in the dataset."""
    # Line 1
    norad_id = str(row["NORAD_CAT_ID"]).rjust(5, "0")
    classification = row["CLASSIFICATION_TYPE"]
    object_id = row["OBJECT_ID"]
    epoch = format_epoch_to_tle(row["EPOCH"])
    mean_motion_dot = f"{row['MEAN_MOTION_DOT']:.8f}"[:10].replace("e", "E").replace("-0", "-").rjust(10, " ")
    mean_motion_ddot = f"{row['MEAN_MOTION_DDOT']:.8f}"[:8].replace("e", "E").replace("-0", "-").rjust(8, " ")
    bstar = f"{row['BSTAR']:.8f}"[:8].replace("e", "E").replace("-0", "-").rjust(8, " ")
    element_set_no = str(row["ELEMENT_SET_NO"]).rjust(4, "0")

    line1 = (f"1 {norad_id}{classification} {object_id.ljust(8)} {epoch} {mean_motion_dot} {mean_motion_ddot} "
             f"{bstar} 0 {element_set_no}")
    line1 = line1[:69] + str(compute_checksum(line1))  # Add checksum

    # Line 2
    inclination = f"{row['INCLINATION']:.4f}".rjust(8, " ")
    ra_of_asc_node = f"{row['RA_OF_ASC_NODE']:.4f}".rjust(8, " ")
    eccentricity = f"{row['ECCENTRICITY']:.7f}"[2:].rjust(7, "0")  # Remove leading "0."
    arg_of_pericenter = f"{row['ARG_OF_PERICENTER']:.4f}".rjust(8, " ")
    mean_anomaly = f"{row['MEAN_ANOMALY']:.4f}".rjust(8, " ")
    mean_motion = f"{row['MEAN_MOTION']:.8f}".rjust(11, " ")
    rev_at_epoch = str(row["REV_AT_EPOCH"]).rjust(5, "0")

    line2 = (f"2 {norad_id} {inclination} {ra_of_asc_node} {eccentricity} {arg_of_pericenter} "
             f"{mean_anomaly} {mean_motion}{rev_at_epoch}")
    line2 = line2[:69] + str(compute_checksum(line2))  # Add checksum

    return {"line1": line1, "line2": line2}

```

```python
# Testing the reconstruction for INSAT-3DS
import pandas as pd
df = pd.read_csv("/content/TLE-data.csv")
insat_row = df[df["OBJECT_NAME"] == "INSAT-3DS"].iloc[0]
tle = reconstruct_tle(insat_row)
print("Reconstructed TLE for INSAT-3DS:")
print(tle["line1"])
print(tle["line2"])
```

applying the reconstruct_tle() function to all rows in the dataset to create the tle_data dict, mapping each OBJECT_NAME to its reconstructed TLE lines

```python
# Create the TLE dictionary
tle_data = {}
for _, row in df.iterrows():
    obj_name = row["OBJECT_NAME"]
    tle_data[obj_name] = reconstruct_tle(row)

# Verify a few entries
print("TLE for INTELSAT 33E DEB:")
print(tle_data["INTELSAT 33E DEB"]["line1"])
print(tle_data["INTELSAT 33E DEB"]["line2"])
```

##calculating 3D distance between two objects using their tles

```python
from skyfield.api import load, EarthSatellite
import numpy as np

ts = load.timescale()
ts = ts.utc(2025,4,18,2)
```

```python
def calculate_distance(obj_name1, obj_name2):
    if obj_name1 not in tle_data or obj_name2 not in tle_data:
        raise ValueError(f"TLE data not found for {obj_name1} or {obj_name2}")

    tle1 = tle_data[obj_name1]
    tle2 = tle_data[obj_name2]

    print(f"Type of t: {type(t)}, Value of t: {t}")  # Debug
    try:
        sat1 = EarthSatellite(tle1["line1"], tle1["line2"], obj_name1, ts)
        sat2 = EarthSatellite(tle2["line1"], tle2["line2"], obj_name2, ts)
        pos1 = sat1.at(t).position.km
        pos2 = sat2.at(t).position.km
        distance = np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2 + (pos1[2] - pos2[2])**2)
        return distance
    except Exception as e:
        print(f"Error calculating distance for {obj_name1} and {obj_name2}: {e}")
        raise
```

```python

pairs = list(itertools.combinations(range(len(df)),2))
X_raw, y = [], []

for i ,j in pairs:
  row1 ,row2 =df.iloc[i], df.iloc[j]
  obj_name1, obj_name2 = row1['OBJECT_NAME'], row2['OBJECT_NAME']
  try:
    distance = calculate_distance(obj_name1, obj_name2)
    print(f"Distance between {obj_name1} and {obj_name2}: {distance:.2f} km")  # Debug
  except ValueError as e:
        print(f"Error for {obj_name1} and {obj_name2}: {e}")
        continue
  # Concatenate features
  features = np.concatenate([row1[base_features].values, row2[base_features].values])
  X_raw.append(features)

  # Risk if distance < 10 km

  y.append(1 if distance < 10 else 0)
  print(f"Added features for {obj_name1} and {obj_name2}, X_raw length: {len(X_raw)}")  # Debug
```

```python
X_raw = np.array(X_raw)
print(f"X_raw shape: {X_raw.shape}, y length: {len(y)}")
y = np.array(y)
```

## preprocessing & scalling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)
```

## **Model Traning**

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"Training set shape: {X_train.shape}, Test set shape: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)


if len(X_scaled) > 0:
    sample = X_scaled[0:1]
    if sample.ndim == 1:
        sample = sample.reshape(1, -1)
    result = {
        "risk": bool(model.predict(sample)[0]),
        "distance_km": calculate_distance(df.iloc[0]['OBJECT_NAME'], df.iloc[1]['OBJECT_NAME']),
        "probability": model.predict_proba(sample)[0][1]
    }
    print(f"Sample Prediction: Risk: {'Yes' if result['risk'] else 'No'}, Distance: {result['distance_km']:.2f} km, Probability: {result['probability']:.2%}")
else:
    print("No data available for prediction.")
```
